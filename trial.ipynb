```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Bandung City Weather Monitoring GKU.csv')
df.head()
```

```python
df.info()
```

```python
df['field6'].plot()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
df = df.set_index('created_at')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='created_at', y='field6', marker='o')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
