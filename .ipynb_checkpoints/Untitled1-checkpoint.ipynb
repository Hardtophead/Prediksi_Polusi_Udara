```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
df = pd.read_csv('polusi.csv')
df.head()
```

```python
df.drop(columns=['latitude', 'longitude', 'elevation', 'status'], inplace=True)
df.head()
```

```python
df.rename(columns={'field1' : 'CO2', 'field2' : 'CO'}, inplace=True)
df.head(3)
```

```python
df.info()
```

```python
# Ubah kolom 'created_at' menjadi tipe datetime
df['created_at'] = pd.to_datetime(df['created_at'])

df.info()
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
for feature in df.select_dtypes(['number']):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```

```python
# Pilih kolom numerik
numeric_df = df.select_dtypes(include=["number"])

# Hitung Q1, Q3, dan IQR
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Filter outliers
condition = ~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)

# Terapkan filter ke DataFrame asli
df_filtered = df.loc[condition]

# Tampilkan hasil
print(df_filtered)
```

```python
sns.histplot(df_filtered['CO2'], kde=True, bins=15)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
 
# Memastikan hanya data dengan tipe numerikal yang akan diproses
numeric_features = df.select_dtypes(include=['number']).columns
numeric_features
 
# Standardisasi fitur numerik
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

```python
sns.histplot(df_filtered['CO'], kde=True, bins=15)
plt.show()
sns.histplot(df_filtered['CO2'], kde=True, bins=15)
plt.show()
```

```python
# Mengidentifikasi baris duplikat
duplicates = df_filtered.duplicated()
 
print("Baris duplikat:")
print(df_filtered[duplicates])
```

```python
# Menghitung jumlah variabel
num_vars = df_filtered.shape[1]
 
# Menentukan jumlah baris dan kolom untuk grid subplot
n_cols = 4  # Jumlah kolom yang diinginkan
n_rows = -(-num_vars // n_cols)  # Ceiling division untuk menentukan jumlah baris
 
# Membuat subplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))
 
# Flatten axes array untuk memudahkan iterasi jika diperlukan
axes = axes.flatten()
 
# Plot setiap variabel
for i, column in enumerate(df_filtered.columns):
    df[column].hist(ax=axes[i], bins=20, edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
 
# Menghapus subplot yang tidak terpakai (jika ada)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
 
# Menyesuaikan layout agar lebih rapi
plt.tight_layout()
plt.show()
```

```python
# Menghitung korelasi antara variabel target dan semua variabel lainnya
target_corr = df_filtered.corr()['CO2']
 
# (Opsional) Mengurutkan hasil korelasi berdasarkan kekuatan korelasi
target_corr_sorted = target_corr.abs().sort_values(ascending=False)
 
plt.figure(figsize=(10, 6))
target_corr_sorted.plot(kind='bar')
plt.title(f'Correlation with CO2')
plt.xlabel('Variables')
plt.ylabel('Correlation Coefficient')
plt.show()
```

```python
import sklearn
from sklearn import datasets
 
# Memisahkan fitur (X) dan target (y)
X = df_filtered.drop(columns=['CO2', 'created_at'])
y = df_filtered['CO2']
 
from sklearn.model_selection import train_test_split
 
# membagi dataset menjadi training dan testing
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
 
# menghitung panjang/jumlah data
print("Jumlah data: ",len(X))
# menghitung panjang/jumlah data pada x_test
print("Jumlah data latih: ",len(x_train))
# menghitung panjang/jumlah data pada x_test
print("Jumlah data test: ",len(x_test))
```

## Lars

```python
from sklearn import linear_model
lars = linear_model.Lars(n_nonzero_coefs=1).fit(x_train, y_train)
 
pred_lars = lars.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae_lars = mean_absolute_error(y_test, pred_lars)
mse_lars = mean_squared_error(y_test, pred_lars)
r2_lars = r2_score(y_test, pred_lars)
 
print(f"MAE: {mae_lars}")
print(f"MSE: {mse_lars}")
print(f"R²: {r2_lars}")
```

```python
# Membuat dictionary untuk menyimpan hasil evaluasi
data = {
    'MAE': [mae_lars],
    'MSE': [mse_lars],
    'R2': [r2_lars]
}
 
# Konversi dictionary menjadi DataFrame
df_results = pd.DataFrame(data, index=['Lars'])
df_results
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression().fit(x_train, y_train)
 
pred_LR = LR.predict(x_test)
 
mae_LR = mean_absolute_error(y_test, pred_LR)
mse_LR = mean_squared_error(y_test, pred_LR)
r2_LR = r2_score(y_test, pred_LR)
 
print(f"MAE: {mae_LR}")
print(f"MSE: {mse_LR}")
print(f"R²: {r2_LR}")
```

```python
df_results.loc['Linear Regression'] = [mae_LR, mse_LR, r2_LR]
df_results
```

## Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
 
GBR = GradientBoostingRegressor(random_state=184)
GBR.fit(x_train, y_train)
 
pred_GBR = GBR.predict(x_test)
 
mae_GBR = mean_absolute_error(y_test, pred_GBR)
mse_GBR = mean_squared_error(y_test, pred_GBR)
r2_GBR = r2_score(y_test, pred_GBR)
 
print(f"MAE: {mae_GBR}")
print(f"MSE: {mse_GBR}")
print(f"R²: {r2_GBR}")
```

```python
df_results.loc['GradientBoostingRegressor'] = [mae_GBR, mse_GBR, r2_GBR]
df_results
```

```python
# Membuat scatter plot menggunakan Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=pred_GBR, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # Line y=x
plt.title("Scatter Plot: Actual vs Predicted (Gradient Boosting Regressor)")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.grid(alpha=0.3)
plt.show()
```
